```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
from sklearn import datasets

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
mouse = merge_df['Mouse ID'].count()
mouse

mouse_unique = merge_df['Mouse ID'].unique()
mouse_count_unique = len(mouse_unique)

print(f'There are ' + str(mouse) + ' mouse records')
print(f'There are ' + str(mouse_count_unique) + ' unique mice')
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Optional: Get all the data for the duplicate mouse ID. 
duplicateRowsDF = merge_df[merge_df.duplicated(['Mouse ID','Timepoint'])]
duplicateRowsDF
```

```python
# Remove duplicate rows based on Mouse ID and Timepoint
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df.drop_duplicates(['Mouse ID','Timepoint'],keep= 'last')
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_clean_list = clean_df['Mouse ID'].count()
mouse_clean_list

mouse_unique_list = merge_df['Mouse ID'].unique()
mouse_count_unique_list = len(mouse_unique_list)

print(f'There are ' + str(mouse_clean_list) + ' mouse records')
print(f'There are ' + str(mouse_count_unique_list) + ' unique mice')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
grouped_drug_df = clean_df.groupby(['Drug Regimen'])

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_vol = grouped_drug_df["Tumor Volume (mm3)"]
mouse_unique_per_drug = grouped_drug_df['Mouse ID'].unique()

tumor_mean = tumor_vol.mean()
tumor_median = tumor_vol.median()
tumor_var = tumor_vol.var()
tumor_std = tumor_vol.std()
tumor_sem = tumor_vol.sem()

# Assemble the resulting series into a single summary dataframe.
drug_summary_df = pd.DataFrame({"Mean": tumor_mean,
                                "Median": tumor_median,
                                "Variance": tumor_var,
                                "Standard Deviation": tumor_std,
                                "SEM": tumor_sem,
                                })
drug_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Create empty list to fill with tumor vol data (for plotting)
vol_data_capomulin_df = drug_filtered_df[drug_filtered_df['Drug Regimen']== 'Capomulin']
vol_data_ramicane_df = drug_filtered_df[drug_filtered_df['Drug Regimen']== 'Ramicane']
vol_data_infubinol_df = drug_filtered_df[drug_filtered_df['Drug Regimen']== 'Infubinol']
vol_data_ceftamin_df = drug_filtered_df[drug_filtered_df['Drug Regimen']== 'Ceftamin']

tum_vol_capomulin_list = vol_data_capomulin_df['Tumor Volume (mm3)'].tolist()
tum_vol_ramicane_list = vol_data_ramicane_df['Tumor Volume (mm3)'].tolist()
tum_vol_infubinol_list = vol_data_infubinol_df['Tumor Volume (mm3)'].tolist()
tum_vol_ceftamin_list = vol_data_ceftamin_df['Tumor Volume (mm3)'].tolist()
```

```python
# IQR analysis Capomulin
quartiles_capomulin = vol_data_capomulin_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
lowerq_capomulin = quartiles_capomulin[0.25]
upperq_capomulin = quartiles_capomulin[0.75]
iqr_capomulin = upperq_capomulin-lowerq_capomulin
lower_bound_capomulin = lowerq_capomulin - (1.5*iqr_capomulin)
upper_bound_capomulin = upperq_capomulin + (1.5*iqr_capomulin)

print(f"The lower quartile of Capomulin Tumor Volume Size is: {lowerq_capomulin}")
print(f"The upper quartile of Capomulin Tumor Volume Size is: {upperq_capomulin}")
print(f"The interquartile range of Capomuin Tumor Volume Size is: {iqr_capomulin}")
print(f"The the median of Capomuolin Tumor Volume Size is: {quartiles_capomulin[0.5]} ")

print(f"Values below {lower_bound_capomulin} could be outliers.")
print(f"Values above {upper_bound_capomulin} could be outliers.")

fig1, ax1 = plt.subplots()
ax1.set_title('Capomulin Final Tumor Size Results')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.boxplot(tum_vol_capomulin_list)
plt.show()

outlier_capomulin = vol_data_capomulin_df.loc[(vol_data_capomulin_df['Tumor Volume (mm3)'] < lower_bound_capomulin) | (vol_data_capomulin_df['Tumor Volume (mm3)'] > upper_bound_capomulin)]
outlier_capomulin
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
tum_vol_capomulin = tumor_avg_by_weight_df['Tumor Volume (mm3)'].tolist()
weight = reset_tumor_avg_by_weight_df['Weight (g)'].tolist()

print(tum_vol_capomulin, weight)

plt.scatter(weight, tum_vol_capomulin, marker="o", facecolors="red", edgecolors="black")
plt.title("Mouse Weight vs. Tumor Volume")
plt.xlabel("Mouse Weight")
plt.ylabel("Average Tumor Volume (mm3)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(weight,tum_vol_capomulin)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
#linear regression
x_values = reset_tumor_avg_by_weight_df['Weight (g)']
y_values = tum_vol_capomulin
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title("Mouse Weight vs. Tumor Volume")
plt.show()
```
